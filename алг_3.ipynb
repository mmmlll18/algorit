```python
from collections import deque

#данные
n = 5
m = 7

edges = [
    (2, 1),
    (3, 2),
    (3, 1),
    (4, 3),
    (4, 1),
    (5, 2),
    (5, 3)
]

#создаем граф и массив входящих степеней
graph = [[] for _ in range(n + 1)]
indegree = [0] * (n + 1)

for u, v in edges:
    graph[u].append(v)
    indegree[v] += 1

#очередь вершин без входящих ребер
q = deque()

for i in range(1, n + 1):
    if indegree[i] == 0:
        q.append(i)

#топологическая сортировка
topological_order = []

while q:
    vertex = q.popleft()
    topological_order.append(vertex)

    for neighbor in graph[vertex]:
        indegree[neighbor] -= 1

        if indegree[neighbor] == 0:
            q.append(neighbor)

#результат
print("Топологический порядок:")
print(*topological_order)
```

```python
#даннне
n = 3
m = 3

edges = [
    (1, 2, 1),
    (2, 3, -3),
    (3, 1, 1)
]

#алгоритм Беллмана-Форда для поиска отрицательного цикла

dist = [0] * (n + 1)

negative_cycle = False

for i in range(n):
    negative_cycle = False

    for u, v, w in edges:
        if dist[v] > dist[u] + w:
            dist[v] = dist[u] + w
            negative_cycle = True

    if not negative_cycle:
        break

print(1 if negative_cycle else 0)
```

Задача сводится к поиску цикла отрицательного веса в ориентированном графе.

Для решения используется алгоритм Беллмана-Форда.

Идея алгоритма:
1. Инициализируем расстояния до всех вершин нулями.
2. Выполняем n итераций релаксации всех ребер.
3. Если на n-й итерации удалось уменьшить хотя бы одно расстояние, то в графе существует цикл отрицательного веса.
4. В этом случае выводим 1.
5. Если изменений нет, выводим 0.

Сложность алгоритма:
O(n · m)

При ограничениях:
n ≤ 1000
m ≤ 10000

алгоритм полностью укладывается в лимиты задачи.

```python
from collections import deque

#данные
n = 3
m = 2

edges = [
    (1, 2),
    (1, 3)
]

graph = [[] for _ in range(n)]
reverse_graph = [[] for _ in range(n)]

for u, v in edges:
    u -= 1
    v -= 1

    graph[u].append(v)
    reverse_graph[v].append(u)

def bfs(start):
    dist = [-1] * n
    q = deque([start])

    dist[start] = 0

    while q:
        v = q.popleft()

        for to in graph[v]:
            if dist[to] == -1:
                dist[to] = dist[v]
                q.append(to)

        for to in reverse_graph[v]:
            if dist[to] == -1:
                dist[to] = dist[v] + 1
                q.append(to)

    return max(dist)

answer = 0

for v in range(n):
    answer = max(answer, bfs(v))

print(answer)
```

Необходимо найти минимальное K, при котором из любого города можно добраться в любой другой, нарушив направление дорог не более K раз.

Построим:
1. Исходный граф дорог.
2. Обратный граф.

При обходе:
- движение по существующей дороге имеет стоимость 0;
- движение против направления дороги считается нарушением и имеет стоимость 1.

Для каждой вершины выполняем поиск в ширину.
Получаем минимальное количество нарушений, необходимое для достижения всех остальных вершин.

Ответом является максимальное значение среди всех найденных расстояний.

Сложность:
Для каждой вершины выполняется обход графа.

O(N · (N + M))

При ограничениях задачи работает достаточно быстро.

```python
# Тестовые данные из примера

procedures = {
    "p1": ["p1", "p2"],
    "p2": ["p1"],
    "p3": ["p1"]
}

names = list(procedures.keys())

# Алгоритм Косарайю

graph = procedures

reverse_graph = {}

for v in names:
    reverse_graph[v] = []

for v in names:
    for to in graph[v]:
        reverse_graph[to].append(v)

used = set()
order = []

def dfs1(v):
    used.add(v)

    for to in graph[v]:
        if to not in used:
            dfs1(to)

    order.append(v)

for v in names:
    if v not in used:
        dfs1(v)

component = {}

def dfs2(v, color):
    component[v] = color

    for to in reverse_graph[v]:
        if to not in component:
            dfs2(to, color)

color = 0

for v in reversed(order):
    if v not in component:
        dfs2(v, color)
        color += 1

# Размер каждой компоненты
sizes = [0] * color

for v in names:
    sizes[component[v]] += 1

# Проверяем потенциальную рекурсивность

for v in names:

    answer = False

    # Компонента содержит цикл
    if sizes[component[v]] > 1:
        answer = True

    # Или есть вызов самой себя
    if v in graph[v]:
        answer = True

    print("YES" if answer else "NO")
```

Процедура является потенциально рекурсивной, если существует путь вызовов, который позволяет вернуться обратно к ней.

Такое возможно тогда и только тогда, когда процедура принадлежит циклу графа вызовов.

Для решения используется алгоритм Косарайю.

1. Строится граф вызовов процедур.
2. Строится обратный граф.
3. Выполняется первый обход DFS для получения порядка обхода вершин.
4. Выполняется второй обход DFS по обратному графу для выделения компонент сильной связности.
5. Если процедура находится в компоненте размером больше 1, то она принадлежит циклу.
6. Также отдельно проверяется случай прямого вызова самой себя.

Для каждой процедуры выводится:
YES — если она потенциально рекурсивна;
NO — если рекурсия невозможна.

```python
import math
from io import StringIO

# данные
input_data = """5
0 0
0 2
1 1
3 0
3 2"""


def read_points(raw_data: str):
    stream = StringIO(raw_data)
    n = int(stream.readline())
    points = []
    for _ in range(n):
        x, y = map(int, stream.readline().split())
        points.append((x, y))
    return points


def distance(p1, p2):
    return math.hypot(p1[0] - p2[0], p1[1] - p2[1])


class DisjointSet:
    def __init__(self, size):
        self.parent = list(range(size))
        self.rank = [0] * size

    def find(self, x):
        while self.parent[x] != x:
            self.parent[x] = self.parent[self.parent[x]]
            # сжатие пути
            x = self.parent[x]
        return x

    def union(self, a, b):
        ra, rb = self.find(a), self.find(b)
        if ra == rb:
            return False
             # a и b уже в одной компоненте - ребро образует цикл
        if self.rank[ra] < self.rank[rb]:
            ra, rb = rb, ra
        self.parent[rb] = ra
        if self.rank[ra] == self.rank[rb]:
            self.rank[ra] += 1
        return True


def minimum_spanning_tree_length(points):

    n = len(points)
    edges = []
    for i in range(n):
        for j in range(i + 1, n):
            edges.append((distance(points[i], points[j]), i, j))
    edges.sort(key=lambda e: e[0])

    dsu = DisjointSet(n)
    total_length = 0.0
    edges_used = 0
    for length, i, j in edges:
        if dsu.union(i, j):
            total_length += length
            edges_used += 1
            if edges_used == n - 1:
                break

    return total_length


points = read_points(input_data)
answer = minimum_spanning_tree_length(points)
print(f"{answer:.9f}")
```



Суть задачи. Нужно соединить n точек на плоскости отрезками так, чтобы
граф получился связным (между любыми двумя точками был путь), а суммарная
длина отрезков — минимальной. Это классическая задача поиска минимального
остовного дерева (MST, Minimum Spanning Tree) на полном графе, где вершины
— заданные точки, а вес ребра между двумя вершинами — евклидово расстояние
между ними.

Алгоритм Крускала. Чтобы построить MST, действуем жадно:
1. Перечисляем все возможные рёбра графа (поскольку любые две точки можно
   соединить отрезком, граф полный — рёбер C(n,2) = n*(n-1)/2) и считаем
   длину каждого ребра по формуле расстояния между точками.
2. Сортируем все рёбра по возрастанию длины.
3. Идём по рёбрам от самого короткого к самому длинному и добавляем ребро
   в остов только если его концы лежат в разных компонентах связности —
   иначе ребро создаст цикл, а циклы в дереве не нужны и не уменьшают вес.
4. Останавливаемся, когда добавлено n-1 рёбер: ровно столько необходимо и
   достаточно, чтобы связать n вершин в дерево без циклов.

Зачем нужна структура Union-Find (DSU). Чтобы быстро проверять, находятся
ли две вершины в одной компоненте связности, и быстро объединять
компоненты при добавлении ребра. Эвристики сжатия пути и объединения по
рангу делают эти операции практически мгновенными, что важно при n до 200
точек (а значит, до ~20000 рёбер).

Почему это даёт минимальный суммарный вес. Доказано (свойство разрезов),
что на каждом шаге самое короткое из ещё не рассмотренных рёбер,
соединяющее две разные компоненты, обязательно входит в какое-нибудь
минимальное остовное дерево. Поэтому жадная стратегия Крускала всегда даёт
именно минимальный суммарный вес.

Формат вывода. Ответ печатается с 9 знаками после запятой, что превышает
требуемую в условии точность 1e-6.

```python
import math
from io import StringIO

#данные
input_data = """12
7 6
4 3
5 1
1 7
2 7
5 7
3 3
7 8
2 8
4 4
6 7
2 6
3"""


def read_input(raw_data: str):

    stream = StringIO(raw_data)
    n = int(stream.readline())
    points = [tuple(map(int, stream.readline().split())) for _ in range(n)]
    k = int(stream.readline())
    return points, k


def distance(p1, p2):

    return math.hypot(p1[0] - p2[0], p1[1] - p2[1])


class DisjointSet:
    def __init__(self, size):
        self.parent = list(range(size))
        self.rank = [0] * size

    def find(self, x):
        while self.parent[x] != x:
            self.parent[x] = self.parent[self.parent[x]]
            x = self.parent[x]
        return x

    def union(self, a, b):
        ra, rb = self.find(a), self.find(b)
        if ra == rb:
            return False
        if self.rank[ra] < self.rank[rb]:
            ra, rb = rb, ra
        self.parent[rb] = ra
        if self.rank[ra] == self.rank[rb]:
            self.rank[ra] += 1
        return True


def max_spacing_clustering(points, k):

    n = len(points)

    edges = []
    for i in range(n):
        for j in range(i + 1, n):
            edges.append((distance(points[i], points[j]), i, j))
    edges.sort(key=lambda e: e[0])

    dsu = DisjointSet(n)
    components = n

    for length, i, j in edges:
        if components == k:
            if dsu.find(i) != dsu.find(j):
                return length
        else:
            if dsu.union(i, j):
                components -= 1

    return 0.0


points, k = read_input(input_data)
answer = max_spacing_clustering(points, k)
print(f"{answer:.9f}")
```


Суть задачи. Даны n точек и число k. Нужно разбить точки на ровно k
непустых групп (кластеров) так, чтобы минимальное расстояние между точками
из разных кластеров (межкластерный разброс) было как можно больше. Это
классическая задача «кластеризации с максимальным разбросом», которая
решается тем же алгоритмом Крускала, что и поиск MST, но с другой
остановкой.

Идея алгоритма. Представим точки как вершины полного графа, где вес ребра
— расстояние между точками. Запустим алгоритм Крускала: рёбра
рассматриваются в порядке возрастания длины, и каждое ребро либо
объединяет две разные компоненты (кластера), либо отбрасывается, если его
концы уже в одном кластере.

Если останавливать этот процесс не на полном дереве (n-1 ребро), а в тот
момент, когда осталось ровно k компонент связности, то мы получаем
разбиение точек на k кластеров способом «ближайшего соседа». Можно
показать, что среди всех возможных разбиений на k кластеров именно такое
разбиение даёт наибольший минимальный разброс между кластерами.

Почему ответ — это вес следующего ребра. Пока количество компонент больше
k, мы объединяем кластеры по самым коротким рёбрам, чтобы получить ровно k
компонент с минимально "разорванными" связями. Как только число компонент
стало равно k, мы перестаём объединять кластеры и ищем первое (самое
короткое из оставшихся) ребро, которое соединяет точки из разных
кластеров. Его длина и есть минимальное расстояние между какими-то двумя
точками из разных кластеров — это и искомый ответ d.

Структура Union-Find используется по той же причине, что и в задаче 18:
быстро объединять кластеры и проверять, лежат ли две точки в одном
кластере.

Граничные случаи. Если k = n, цикл сразу попадёт в ветку "components ==
k" и вернёт длину самого короткого ребра среди всех точек — это и есть
правильный ответ.
