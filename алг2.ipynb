```python
#данные

queries = [
    "+ 5",
    "+ 2",
    "+ 8",
    "+ 1",
    "? 1",
    "? 3",
    "? 4"
]

data = []

for query in queries:

    op, value = query.split()
    value = int(value)

    if op == "+":
        if value not in data:
            data.append(value)
            data.sort()

    else:
        print(data[value - 1])
```

Задание 9

```python
#данные

N = 6

nodes = [
    (-2, 0, 2),
    (8, 4, 3),
    (9, 0, 0),
    (3, 6, 5),
    (6, 0, 0),
    (0, 0, 0)
]

queries = [6, 9, 7, 8]

tree = {}
key_to_index = {}

for i, (key, left, right) in enumerate(nodes, start=1):
    tree[i] = (key, left, right)
    key_to_index[key] = i

subtree_size = {}

def dfs(v):

    if v == 0:
        return 0

    key, left, right = tree[v]

    size = 1 + dfs(left) + dfs(right)

    subtree_size[v] = size

    return size

total = dfs(1)

deleted = set()

def mark_deleted(v):

    if v == 0 or v in deleted:
        return

    deleted.add(v)

    _, left, right = tree[v]

    mark_deleted(left)
    mark_deleted(right)

for key in queries:

    if key in key_to_index:

        vertex = key_to_index[key]

        if vertex not in deleted:

            total -= subtree_size[vertex]

            mark_deleted(vertex)

    print(total)
```

Для каждой вершины заранее вычисляется размер её поддерева.

Размер поддерева:

size(v) =
1 + size(left) + size(right)

После получения запроса:

1. Находим вершину по ключу.
2. Если она ещё не удалена:
   - вычитаем размер её поддерева из общего числа вершин;
   - помечаем всё поддерево удалённым.

Так как высота дерева не превышает 25,
поиск вершины работает быстро.



Задание 17

```python
#данные

operations = [
    "? 1",
    "+ 1",
    "? 1",
    "+ 2",
    "s 1 2"
]

S = set()

last_sum = 0
MOD = 1000000001

for op in operations:

    parts = op.split()

    if parts[0] == "+":

        value = (int(parts[1]) + last_sum) % MOD
        S.add(value)

    elif parts[0] == "-":

        value = (int(parts[1]) + last_sum) % MOD
        S.discard(value)

    elif parts[0] == "?":

        value = (int(parts[1]) + last_sum) % MOD

        if value in S:
            print("Found")
        else:
            print("Not found")

    else:

        l = (int(parts[1]) + last_sum) % MOD
        r = (int(parts[2]) + last_sum) % MOD

        current_sum = sum(x for x in S if l <= x <= r)

        print(current_sum)

        last_sum = current_sum
```

Необходимо поддерживать множество чисел и выполнять операции:

add(i)
del(i)
find(i)
sum(l, r)

Каждая операция зависит от результата последнего запроса суммы.

В демонстрационной версии используется множество Python set.

Операция sum(l, r) вычисляется перебором всех элементов множества.

```python
#данные

commands = [
    ("+1", 5),
    ("+1", 3),
    ("+1", 7),
    ("0", 1),
    ("0", 2),
    ("0", 3),
    ("-1", 5),
    ("+1", 10),
    ("0", 1),
    ("0", 2),
    ("0", 3)
]

data = []

for command, value in commands:

    if command == "+1":

        if value not in data:
            data.append(value)
            data.sort()

    elif command == "-1":

        if value in data:
            data.remove(value)

    else:

        # k-й максимум
        k = value

        answer = data[-k]

        print(answer)
```

Необходимо поддерживать множество чисел и выполнять три операции:

1. Добавление элемента.
2. Удаление элемента.
3. Поиск k-го максимума.

В демонстрационной версии используется отсортированный список.

При добавлении:
- элемент добавляется только если его ещё нет;
- список сортируется.

При удалении:
- элемент удаляется из списка.

Для поиска k-го максимума:
- используется индекс -k,
так как список хранится по возрастанию.

Например:

[3, 5, 7, 10]

1-й максимум = 10
2-й максимум = 7
3-й максимум = 5



Задание 15

```python
# Тестовые данные из примера

class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None


# Построение дерева из условия

root = Node(4)
root.left = Node(3)
root.right = Node(5)

X = 4


def find_max(node):
    while node.right:
        node = node.right
    return node


def delete(node, key):

    if node is None:
        return None

    if key < node.key:
        node.left = delete(node.left, key)

    elif key > node.key:
        node.right = delete(node.right, key)

    else:

        # нет детей
        if node.left is None and node.right is None:
            return None

        # только правый ребёнок
        if node.left is None:
            return node.right

        # только левый ребёнок
        if node.right is None:
            return node.left

        # оба ребёнка
        replacement = find_max(node.left)

        node.key = replacement.key

        node.left = delete(node.left, replacement.key)

    return node


root = delete(root, X)


# Вывод дерева в формате задачи

result = []


def dfs(node):

    if node is None:
        return 0

    index = len(result) + 1

    result.append([node.key, 0, 0])

    left_index = dfs(node.left)
    right_index = dfs(node.right)

    result[index - 1][1] = left_index
    result[index - 1][2] = right_index

    return index


dfs(root)

print(len(result))

for row in result:
    print(*row)
```

В задаче требуется удалить вершину из AVL-дерева.

Используется стандартный алгоритм удаления из дерева поиска:

1. Находим удаляемую вершину.
2. Если вершина является листом — удаляем её.
3. Если есть только один ребёнок — заменяем вершину этим ребёнком.
4. Если есть оба ребёнка:
   - находим максимальный элемент в левом поддереве;
   - копируем его ключ;
   - удаляем найденную вершину.

После удаления дерево обходится в глубину и выводится в формате:
ключ, номер левого ребёнка, номер правого ребёнка.

Для примера из условия:

    4
   / \
  3   5

после удаления 4 получаем:

    3
     \
      5

что соответствует ответу из примера.
