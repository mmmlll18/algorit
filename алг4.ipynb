Задание 3


```python
#данные взятые из примера

pattern = "aba"
text = "abacaba"

p_len = len(pattern)
t_len = len(text)

base = 257
mod = 1000000007

pattern_hash = 0
current_hash = 0

power = 1

for i in range(p_len):

    pattern_hash = (pattern_hash * base + ord(pattern[i])) % mod
    current_hash = (current_hash * base + ord(text[i])) % mod

    if i < p_len - 1:
        power = (power * base) % mod

positions = []

for i in range(t_len - p_len + 1):

    if pattern_hash == current_hash:

        if text[i:i + p_len] == pattern:
            positions.append(i + 1)

    if i < t_len - p_len:

        current_hash = (
            current_hash
            - ord(text[i]) * power
        ) % mod

        current_hash = (
            current_hash * base
            + ord(text[i + p_len])
        ) % mod

print("Количество вхождений:")
print(len(positions))

print("Позиции:")
print(*positions)
```

Используется алгоритм Рабина-Карпа.

Для строки-паттерна и каждого окна текста одинаковой длины вычисляется полиномиальный хеш.

Если хеши совпали, дополнительно выполняется проверка через оператор == для исключения коллизий.

Для быстрого пересчета используется скользящий хеш:

1. Удаляется вклад первого символа окна.
2. Хеш сдвигается.
3. Добавляется новый символ.

Таким образом каждое новое окно обрабатывается за O(1).


```python
# данные

s = "abacaba"

n = len(s)

z = [0] * n

left = 0
right = 0

for i in range(1, n):

    if i <= right:
        z[i] = min(
            right - i + 1,
            z[i - left]
        )

    while (
        i + z[i] < n and
        s[z[i]] == s[i + z[i]]
    ):
        z[i] += 1

    if i + z[i] - 1 > right:
        left = i
        right = i + z[i] - 1

print(*z[1:])
```

Z-функция строки показывает для каждой позиции i длину наибольшего префикса строки, совпадающего с подстрокой, начинающейся в позиции i.

Для вычисления используется линейный алгоритм.

Поддерживается текущий отрезок [left, right], внутри которого уже известно совпадение с префиксом.

Если новая позиция находится внутри этого отрезка, часть значения Z можно взять сразу.

Затем значение увеличивается посимвольным сравнением.

После каждого расширения границы отрезка обновляются.

Сложность алгоритма:

Время: O(n)
Память: O(n)

где n — длина строки.

```python
#данные из примера
s = "ABCABCDEDEDEF"

n = len(s)

# dp[i][j]-оптимальное представление подстроки s[i:j+1]
dp = [["" for _ in range(n)] for _ in range(n)]

for length in range(1, n + 1):

    for i in range(n - length + 1):

        j = i + length - 1

        current = s[i:j + 1]
        best = current

        for k in range(i, j):

            candidate = dp[i][k] + "+" + dp[k + 1][j]

            if len(candidate) < len(best):
                best = candidate

        # Проверка на повторяющийся шаблон
        substring = current
        L = len(substring)

        for part_len in range(1, L):

            if L % part_len != 0:
                continue

            pattern = substring[:part_len]

            if pattern * (L // part_len) == substring:

                pattern_repr = dp[i][i + part_len - 1]

                times = L // part_len

                if times > 1:
                    candidate = f"{pattern_repr}*{times}"

                    if len(candidate) < len(best):
                        best = candidate

        dp[i][j] = best

print("Исходная строка:")
print(s)

print("\nОптимальное представление:")
print(dp[0][n - 1])
```



В данной задаче необходимо представить строку в наиболее компактном виде.

Допускаются две операции:

1. Конкатенация частей строки с помощью знака "+".
2. Представление повторяющейся подстроки в виде:
   строка*количество.

Например:

ABCABC → ABC*2
DEDEDE → DE*3

Для поиска оптимального представления используется динамическое программирование.

Для каждой подстроки рассматриваются три варианта:

• оставить подстроку без изменений;
• разбить её на две части и объединить оптимальные представления этих частей;
• представить её как повторение некоторого шаблона.

Из всех вариантов выбирается запись минимальной длины.

Для строки:

ABCABCDEDEDEF

программа получает результат:

ABC*2+DE*3+F

что соответствует примеру из условия.

```python
#примерные данные

s = "trololo"

queries = [
    (0, 0, 7),
    (2, 4, 3),
    (3, 5, 1),
    (1, 3, 2)
]

n = len(s)

MOD1 = 1000000007
MOD2 = 1000000009
X = 911382323

h1 = [0] * (n + 1)
h2 = [0] * (n + 1)

pow1 = [1] * (n + 1)
pow2 = [1] * (n + 1)

for i in range(n):

    h1[i + 1] = (h1[i] * X + ord(s[i])) % MOD1
    h2[i + 1] = (h2[i] * X + ord(s[i])) % MOD2

    pow1[i + 1] = (pow1[i] * X) % MOD1
    pow2[i + 1] = (pow2[i] * X) % MOD2


def get_hash(left, length):

    hash1 = (
        h1[left + length]
        - h1[left] * pow1[length]
    ) % MOD1

    hash2 = (
        h2[left + length]
        - h2[left] * pow2[length]
    ) % MOD2

    return hash1, hash2


for a, b, l in queries:

    if get_hash(a, l) == get_hash(b, l):
        print("Yes")
    else:
        print("No")
```

Для решения используется двойное полиномиальное хеширование.

Предварительно вычисляются:

1. Хеши всех префиксов строки.
2. Степени основания X.

После этого хеш любой подстроки можно получить за O(1):

H(l, r) =
h[r] − h[l] · X^(r−l)

Для уменьшения вероятности коллизий используются два модуля:

MOD1 = 10^9 + 7
MOD2 = 10^9 + 9

Подстроки считаются равными только если совпадают оба хеша.



Задание 8

```python
#двнные

k = 1
text = "ababab"
pattern = "baaa"

n = len(text)
m = len(pattern)

MOD = 1000000007
BASE = 257

max_len = max(n, m)

power = [1] * (max_len + 1)

for i in range(1, max_len + 1):
    power[i] = (power[i - 1] * BASE) % MOD


def build_hash(string):

    h = [0] * (len(string) + 1)

    for i in range(len(string)):
        h[i + 1] = (
            h[i] * BASE + ord(string[i])
        ) % MOD

    return h


ht = build_hash(text)
hp = build_hash(pattern)


def get_hash(h, left, length):

    return (
        h[left + length]
        - h[left] * power[length]
    ) % MOD


positions = []

for start in range(n - m + 1):

    mismatches = 0
    pos = 0

    while pos < m and mismatches <= k:

        left = pos
        right = m - pos

        #бинарный поиск первой ошибки

        while right > 0:

            mid = right // 2

            if get_hash(ht, start + left, mid + 1) == \
               get_hash(hp, left, mid + 1):

                left += mid + 1
                right -= mid + 1

            else:
                right = mid

        if left >= m:
            break

        mismatches += 1
        pos = left + 1

    if mismatches <= k:
        positions.append(start)

print(len(positions), *positions)
```

Требуется найти все позиции, где паттерн отличается от подстроки текста не более чем на k символов.

Для быстрого сравнения подстрок используются полиномиальные хеши.

Для каждой позиции текста:

1. Сравнивается подстрока текста и паттерн.
2. Если они различаются, необходимо найти первое несовпадение.
3. Первое несовпадение ищется бинарным поиском по хешам.
4. После нахождения ошибки продолжаем поиск дальше.
5. Разрешается не более k несовпадений.

Так как k ≤ 5, количество бинарных поисков очень маленькое.



Задание "ИМЕНА"

```python
# Исходные данные из примера

X = "abcabca"
Y = "abcda"

n = len(X)
m = len(Y)

dp = [["" for _ in range(m + 1)] for _ in range(n + 1)]

for i in range(n - 1, -1, -1):
    for j in range(m - 1, -1, -1):

        best = ""

        if X[i] == Y[j]:
            best = X[i] + dp[i + 1][j + 1]

        if dp[i + 1][j] > best:
            best = dp[i + 1][j]

        if dp[i][j + 1] > best:
            best = dp[i][j + 1]

        dp[i][j] = best

print("Имя ребенка:")
print(dp[0][0])
```

В данной задаче необходимо определить лексикографически наибольшее имя ребёнка, которое можно получить одновременно из имени отца и имени матери удалением некоторых символов. При этом порядок оставшихся символов должен сохраняться.

Таким образом, искомое имя должно являться общей подпоследовательностью двух строк.

Для решения задачи используется метод динамического программирования.

Создаётся таблица dp, где элемент dp[i][j] хранит наилучшую (лексикографически максимальную) подпоследовательность, которую можно получить из суффиксов строк X[i:] и Y[j:].

При заполнении таблицы рассматриваются следующие варианты:

Если текущие символы строк совпадают, то данный символ может быть включён в ответ, после чего продолжается поиск для следующих позиций.
Можно пропустить текущий символ строки отца и перейти к следующему символу.
Можно пропустить текущий символ строки матери и перейти к следующему символу.

Из всех полученных вариантов выбирается строка, которая является лексикографически наибольшей.

Лексикографическое сравнение выполняется аналогично сравнению слов в словаре. Например:

z > y
cab > caa
ca > bzzz

После заполнения таблицы ответом является значение dp[0][0], то есть лучшее имя, которое можно получить из исходных строк.



Задача ДНК

```python
#данные

start = list("AGCT")
target = list("GCAT")

operations = []

n = len(start)

for i in range(n):

    if start[i] == target[i]:
        continue

    pos = i

    while pos < n and start[pos] != target[i]:
        pos += 1

    start[i:pos + 1] = reversed(start[i:pos + 1])

    operations.append((i + 1, pos + 1))

print(len(operations))

for l, r in operations:
    print(l, r)
```

В данной задаче необходимо преобразовать исходную цепочку ДНК в требуемую цепочку.

Каждая цепочка представляется строкой, состоящей из символов:

A, G, C, T

Разрешённая операция заключается в развороте произвольного непрерывного участка строки между позициями L и R.

Например:

AGCT

после разворота участка от 2 до 4 превращается в:

ATCG

Для решения задачи используется жадный алгоритм.

Преобразование выполняется последовательно слева направо.

Для каждой позиции строки:

Проверяется, совпадает ли текущий символ с требуемым символом в целевой строке.
Если символ уже находится на нужном месте, программа переходит к следующей позиции.
Если символ отличается, то справа ищется нужный символ.
После нахождения нужного символа выполняется разворот отрезка от текущей позиции до найденного символа.
В результате нужный символ оказывается на своей позиции.

После этого алгоритм переходит к следующей позиции и продолжает работу до конца строки.

Таким образом, после обработки всех позиций исходная строка постепенно преобразуется в требуемую.


Первое число обозначает количество выполненных преобразований, а каждая последующая строка содержит границы разворачиваемого участка.

После выполнения всех операций исходная цепочка полностью совпадает с требуемой цепочкой ДНК.
