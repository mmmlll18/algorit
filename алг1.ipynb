Задание 4


```python
#данные
segments = [
    (1, 3),
    (2, 5),
    (3, 6)
]

#сортируем отрезки по правому концу
segments.sort(key=lambda x: x[1])

points = []

current_point = segments[0][1]
points.append(current_point)

for left, right in segments[1:]:
    if left > current_point:
        current_point = right
        points.append(current_point)

print("Минимальное количество точек:")
print(len(points))

print("Координаты точек:")
print(*points)
```

Используется жадный алгоритм.

1. Все отрезки сортируются по правой границе.
2. Берется правая граница первого отрезка и добавляется в ответ.
3. Далее последовательно рассматриваются остальные отрезки.
4. Если текущая выбранная точка не принадлежит очередному отрезку, выбирается его правая граница.
5. Таким образом одной точкой покрывается максимально возможное число отрезков.

Доказано, что такой жадный выбор всегда дает минимальное количество точек.



Задание 7

```python
#данные

K = 10
n = 3

times = [6, 2, 8]

#сортируем время ремонта по возрастанию
times.sort()

repaired = 0
spent_time = 0

for t in times:
    if spent_time + t <= K:
        spent_time += t
        repaired += 1
    else:
        break

print("Максимальное количество сапог:")
print(repaired)
```

Необходимо починить максимальное количество сапог за ограниченное время K.

Для этого используется жадный алгоритм:

1. Сортируем все времена ремонта по возрастанию.
2. Сначала ремонтируем самые быстрые сапоги.
3. Пока суммарное время не превышает K, добавляем очередной ремонт.
4. Как только следующий сапог не помещается в оставшееся время, завершаем работу.

Почему алгоритм правильный:
если заменить более долгий ремонт более коротким, количество отремонтированных сапог не уменьшится, поэтому всегда выгодно сначала брать самые быстрые работы.



Задание 11

```python
#данные

W = 10
weights = [1, 4, 8]

#DP[i] = максимальный вес золота,
#который можно набрать при вместимости i

dp = [0] * (W + 1)

for weight in weights:
    for capacity in range(W, weight - 1, -1):
        dp[capacity] = max(
            dp[capacity],
            dp[capacity - weight] + weight
        )

print("Максимальный вес золота:")
print(dp[W])
```

Используется динамическое программирование для задачи 0/1 рюкзака.

Пусть dp[i] — максимальный вес золота, который можно набрать при вместимости рюкзака i.

Для каждого слитка веса weight:
1. Перебираем вместимости от W до weight.
2. Решаем:
   - не брать слиток;
   - взять слиток.

Переход:

dp[i] = max(dp[i], dp[i - weight] + weight)

Обратный обход необходим, чтобы каждый слиток использовался не более одного раза.



Задание 15

```python
# Тестовые данные

s = "([)]"

n = len(s)

dp = [[0] * n for _ in range(n)]
parent = [[None] * n for _ in range(n)]

pairs = {
    '(': ')',
    '[': ']',
    '{': '}'
}

for length in range(2, n + 1):
    for l in range(n - length + 1):
        r = l + length - 1

        dp[l][r] = 10**9

        if s[l] in pairs and pairs[s[l]] == s[r]:
            value = 0 if l + 1 > r - 1 else dp[l + 1][r - 1]

            if value < dp[l][r]:
                dp[l][r] = value
                parent[l][r] = (-1, -1)

        for k in range(l, r):
            value = dp[l][k] + dp[k + 1][r]

            if value < dp[l][r]:
                dp[l][r] = value
                parent[l][r] = (l, k)

        dp[l][r] += 1


def restore(l, r):
    if l > r:
        return ""

    if l == r:
        return ""

    p = parent[l][r]

    if p == (-1, -1):
        return s[l] + restore(l + 1, r - 1) + s[r]

    if p is not None:
        _, k = p
        return restore(l, k) + restore(k + 1, r)

    return ""


print("Полученная правильная последовательность:")
print(restore(0, n - 1))
```

Используется динамическое программирование по подстрокам.

dp[l][r] хранит минимальное количество удалений,
необходимое для превращения подстроки s[l..r]
в правильную скобочную последовательность.

Для каждой подстроки рассматриваются два варианта:

1. Крайние скобки образуют правильную пару.
2. Подстрока разбивается на две части.

Задание 20

```python
#данные из примерв

N = 5
K = 1
S = "abcde"

#DP для количества несовпадающих пар

dp = [[0] * N for _ in range(N)]

answer = 0

# подстроки длины 1
for i in range(N):
    answer += 1

#подстроки длины >= 2
for length in range(2, N + 1):

    for left in range(N - length + 1):

        right = left + length - 1

        if length == 2:
            dp[left][right] = (
                0 if S[left] == S[right] else 1
            )
        else:
            dp[left][right] = (
                dp[left + 1][right - 1]
                + (0 if S[left] == S[right] else 1)
            )

        if dp[left][right] <= K:
            answer += 1

print("Количество почти палиндромов:")
print(answer)
```

Используется динамическое программирование по подстрокам.

Пусть dp[l][r] — количество несовпадающих пар символов в подстроке S[l..r].

Если крайние символы совпадают, то число необходимых замен не увеличивается.

Если крайние символы различны, потребуется одна замена.

Переход:

dp[l][r] =
dp[l+1][r-1] + (S[l] != S[r])

Подстрока является почти палиндромом, если количество необходимых замен не превышает K.

После вычисления dp для всех подстрок подсчитывается количество подстрок, удовлетворяющих условию.
